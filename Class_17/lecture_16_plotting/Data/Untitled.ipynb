```python
import pandas as pd

df_mtcars = pd.read_csv("mtcars.csv")
df_mtcars.head(10)
```

```python
df_mtcars.columns = df_mtcars.columns.\
                        str.replace("_", " ")
    
df_mtcars["Engine"] = df_mtcars.vs.map({0:"Straight", 1:"L-shaped"})
```

```python
df_mtcars.columns = df_mtcars.columns.str.replace(" ","_")
```

```python
def Mazda_Check(car_name):
    
    if "Mazda" in car_name:
        return "Mazda"
    else:
        return "Other"


df_mtcars["Mazda?"] = df_mtcars.car_name.apply(Mazda_Check)
```

```python
df_mtcars.head()
```

```python
#cool = mpg>=20, hp >=150
def Get_Cool(row):
    
    hp = row["hp"]
    mpg = row["mpg"]
    if hp>=150 and mpg>=15:
        return "cool"
    else:
        return "Other"


df_mtcars["Cool?"] = df_mtcars[["mpg", "hp"]].apply(Get_Cool, axis = 1)
```
