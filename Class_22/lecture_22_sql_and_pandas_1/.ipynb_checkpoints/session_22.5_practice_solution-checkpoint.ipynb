# Yelp Data Set

In this series of practice problems we will explore the data set in "Yelp_Data.csv", which contains data from Yelp on various restaurants across the US. Our goal is to find every states favorite food among the following 4 categories:

- Pizza
- Mexican
- Burgers
- Chinese

Read in the dataset with na_values = ["NA"].  Then drop all rows (axis=0) with an NaN .

```python
import pandas as pd
from pandasql import sqldf

#Write you code here
df_yelp = pd.read_csv("Data/Yelp_Data.csv", na_values=["NA"])
df_yelp.dropna(axis=0, inplace=True)

df_yelp.head()
```

Let's only look at states with at least 500 restaurants. Write an SQL query to find these states.

```python
#Write your query to create df_yelp_2 below
key_states = sqldf("SELECT state, COUNT(state) as total_restuarants FROM df_yelp GROUP BY state HAVING total_restuarants>=500")
key_states
```

Create a new dataframe called df_yelp_2, which only has data from restaurants in these states

```python
df_yelp_2 = sqldf("SELECT * FROM df_yelp WHERE state IN ('AZ', 'NV', 'NC', 'PA', ' EDH', 'WI', 'QC', 'IL', 'BW') ")

df_yelp_2.head()
```

Now create df_yelp_3 that only stores restaurants that identify as one of the four types of restaurants we care about.

```python
#Write your code here
df_yelp_3 = sqldf("SELECT * FROM df_yelp_2\
                    WHERE (Pizza + Mexican + Burgers + Chinese) = 1  ")

df_yelp_3.head()
```

Use SQL to greate df_yelp_4 that is the same as df_yelp_3 with a column "Restaurant_Type" that simply identifies the restaurant type.

```python
#Write you code here
df_yelp_4 = sqldf("SELECT *, CASE WHEN Mexican =1 THEN 'Mexican' WHEN Pizza=1 THEN 'Pizza' WHEN Burgers=1 THEN 'Burger' ELSE 'Chinese' END AS Restaurant_Type FROM df_yelp_3 ")

df_yelp_4.head()
```

Next create a column called "Total_Stars" that is stars*review_count. You can just use pandas for this one

```python
#Write your code here
df_yelp_4["Total_Stars"] = df_yelp_4["review_count"]*df_yelp_4["stars"]

df_yelp_4.head()
```

For each restaurant type and each state compute the sum of total_stars and review_count.  Use SQL and store result in a dataframe called df_yelp_5

```python
#Write your code here
df_yelp_5 = sqldf("SELECT state, Restaurant_Type, SUM(Total_Stars) AS Sum_Stars, SUM(review_count) Total_Reviews FROM df_yelp_4\
                    GROUP BY state, Restaurant_Type")

df_yelp_5
```

Add a column that is called "Avg_Stars" = "Sum_Stars"/"Total_Reviews" to df_yelp_5 using pandas.

```python
#Write your code here
df_yelp_5["Avg_Stars"] = df_yelp_5["Sum_Stars"]/df_yelp_5["Total_Reviews"]

df_yelp_5
```

Write one final query to find every state's favorite restaurant type out of the four proposed options.

```python
final_result = sqldf("SELECT state, Restaurant_Type, MAX(Avg_Stars) As Max_Stars FROM df_yelp_5 GROUP BY state")

final_result
```
